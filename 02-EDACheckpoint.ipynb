##EDA Relevance

The goal of this EDA is to examine whether host countries tend to achieve unusually strong medal performances in the years they host the Olympic Games. To investigate this question, I focused on all Summer Olympic host nations from 1996 to 2016 (United States, Australia, Greece, China, Great Britain, and Brazil) and tracked each country’s total medal counts across all Olympic cycles during this period.

By visualizing each country individually, I am able to compare their host-year performance against their own historical baseline rather than against other nations with very different competitive strengths. This approach directly relates to the research question because it isolates the “host effect” within each country’s overall performance trend and highlights whether hosting corresponds to a notable increase in medal count.

Overall, the figures included here allow me to evaluate whether host years represent a meaningful deviation from typical performance for each nation, making this EDA directly relevant to answering the central question of host advantages in Olympic outcomes.

##EDA Analysis and Description

Each subplot displays a single country’s medal trajectory from 1996 to 2016, with the host year highlighted in red. Across the six countries, several consistent patterns emerge. Nations such as Australia (2000), China (2008), Great Britain (2012), and Brazil (2016) show a clear local peak in medal counts during their host year. These spikes stand out from the surrounding years and suggest that hosting may provide a competitive advantage, possibly due to increased funding, national motivation, or familiar environmental conditions.

In contrast, Greece (2004) shows only a modest increase, and the United States (1996) shows no dramatic deviation from its already high performance level, indicating that the magnitude of the host effect may depend on the underlying strength and consistency of the country’s athletic programs.

The individual-country figures also reveal broader performance trends beyond the host effect. For example, Great Britain shows a strong upward trajectory leading into 2012, suggesting long-term investment rather than a single-year spike. Brazil’s pattern is the opposite: its 2016 host performance is relatively strong compared to surrounding years, making the host boost more visually apparent.

Overall, these visual analyses indicate that while hosting often coincides with improved medal outcomes, the strength of this effect varies by country and depends on their long-term performance trends.

```python
## YOUR CODE HERE
import numpy as np

focus_countries = list(host_country_1996_2016.values())

focus_df = country_year_medals[
    (country_year_medals['Team'].isin(focus_countries)) &
    (country_year_medals['Year'].between(1996, 2016))
]
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(20, 10))
axes = axes.flatten()

for idx, country in enumerate(focus_countries):
    ax = axes[idx]
    temp = focus_df[focus_df['Team'] == country]

    # Plot line
    ax.plot(temp['Year'], temp['MedalCount'], marker='o', linewidth=2, markersize=8)

    # Mark host year
    for year, host in host_country_1996_2016.items():
        if host == country:
            host_row = temp[temp['Year'] == year]
            if not host_row.empty:
                host_medals = host_row['MedalCount'].iloc[0]
                ax.scatter(year, host_medals, color='red', s=120, zorder=5)

                # shift label upward for clarity
                ax.text(year, host_medals + 3, 'HOST', color='red', fontsize=12, ha='center', fontweight='bold')

    # Titles & styling
    ax.set_title(f'{country}', fontsize=16, fontweight='bold')
    ax.set_xlabel('Year', fontsize=12)
    ax.set_ylabel('Medals', fontsize=12)
    ax.grid(True, alpha=0.3)

# Adjust the layout so titles and labels don’t overlap
plt.tight_layout()
plt.suptitle('Host Countries Medal Trends (1996–2016)', fontsize=20, fontweight='bold', y=1.03)
plt.show()
```
